## 코사인 유사도(Cosine similarity)
---
$$similarity = cos(\theta) = \displaystyle\frac{A \cdot B}{{\lvert\lvert A \rvert\rvert}\;{\lvert\lvert B \rvert\rvert}} $$

```python
# 코사인 유사도
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))

doc1 = np.array([0,1,1,1])
doc2 = np.array([1,0,1,1])
doc3 = np.array([2,0,2,2])

print("doc1 doc2 유사도: ", cos_sim(doc1, doc2))
print("doc2 doc3 유사도: ", cos_sim(doc2, doc3))
print("doc1 doc3 유사도: ", cos_sim(doc1, doc3))
```

#### doc2, doc3의 유사도가 1인 이유
- 두 벡터의 단위벡터가 동일하기 때문
#### doc1, doc2와 doc1, doc3의 유사도가 동일한 이유
- doc2, doc3의 유사도가 1인 이유와 동일
