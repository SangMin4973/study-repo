## 시퀀스투시퀀스(Sequence-to-Sequence, seq2seq)
- 입력된 시퀀스로부터 다른 도메인의 시퀀스를 출력하는 모델

#### 모델의 개요
- 

- Encoder -> Context Vector -> Decoder 구조
- Context Vector: 인코더 RNN셀의 마지막 시점의 은닉 상태

#### seq2seq의 동작 과정


- 테스트 단계
    - 디코더 첫 문장의 시작 \<sos>(start of string)가 들어감
    - \<sos> 가 들어오면 다음에 등잘할 확률이 높은 단어 예측
    - 다음 입력으로는 전에 예측한 단어가 들어감
    - \<eos>(end of string)가 다음 단어로 예측될 때 까지 반복

- Training
    - 교사 강요: 학습시 다음 스텝에 들어가는 입력값이 기존 RNN은 이전 단계의 출력을 입력으로 사용하는 반면 교사 강요는 정답 레이블이 들어가게 된다. 

- Embedding Layer
    - 텍스트를 벡터로 변환하는 층

- RNN cell
    - RNN 셀에서는 현재 시점의 입력과 이전 시점의 출력(은닉 상태)을 같이 입력받는다.
    - 마지막 셀의 은닉 상태는 문맥 벡터(Context Vector)가 된다.

- Decoder

    - 

    - \<sos>를 시작으로 다음에 올 단어들을 예측한다.
