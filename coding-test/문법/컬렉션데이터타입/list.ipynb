#### 뮤터블 객체(mutable object)
- 데이터 수정이 가능한 객체
- 리스트, 딕셔너리, 셋

리스트

```python
my_list = [1,2,3,4,5]
my_list[4] = 6
my_list
```

```python
my_list2 = [1,3,5] + [7,9]
print(my_list2)

my_list3 = list(my_list)
print(my_list)
print(my_list2)
print(my_list3)
```

```python
# 리스트 인덱싱
my_list = [1,2,4]
my_list.append(6)
print(my_list, my_list[2])

# del로 인덱스 위치 원소 지우기
del my_list[2]
print(my_list)
```

```python
# 리스트 슬라이싱
my_list = [1,2,3,4,5]
print(my_list[0:2])
print(my_list[1:])
print(my_list[3:4])
print(my_list[-4:-2])
```

```python
# 리스트 컴프리헨션
numbers = [x for x in range(5)]
numbers
```

```python
cartesian_product = [(x, y) for x in range(2) for y in range(3)]
cartesian_product
```

```python
def square(x):
    return x*x

squares = [square(x) for x in range(5)]
squares
```

```python
# map()과 람다식의 조합
squares = list(map(lambda x: x*x, range(5)))
squares
```

```python
# 조건문이 있는 리스트 컴프리헨션
even_numbers = [x for x in range(5) if x%2 == 0]
even_numbers
```
