#### 셋(set)
- 해시 테이블을 기반으로 한 자료형
- 중복 허용 x
- 집합 연산을 제공

```python
empty_set = set()
empty_set
```

```python
# 리스트를 셋으로 변환
list_to_set = set([1,2,3,3,2])
list_to_set
```

```python
# 중괄호를 이용한 셋 생성
set_from_braces = {12,14,1,7,7}
set_from_braces
```

```python
# 문자열을 셋으로 변환
word = "hello"
string_to_set = set(word)
string_to_set
```

```python
# 튜플을 셋으로 변환
tuple_to_set = set((1,2,3,3))
tuple_to_set
```

```python
# 셋 컴프리헨션
comprehension_set = {x for x in range(5) if x % 2 == 0}
comprehension_set
```

```python
# 셋에 원소 추가
my_set = {1,2,3}
my_set.add(4)
my_set
```

```python
# 셋에 여러 원소 추가
my_set.update([3,4,5,6,7])
my_set
```

```python
# 셋에 원소 제거
my_set.remove(2)
my_set
```

```python
# my_set.remove(10) -> 오류 발생(없는 원소)
my_set.discard(10) # 없는 원소여도 오류 발생 x
my_set
```

```python
# 모든 원소 제거
my_set.clear()
my_set
```

```python
# 셋의 합집합
set1 = {1,2,3}
set2 = {3,4,5}
union_set = set1.union(set2)
union_set
```

```python
set1 |= set2 #union과 동일
set1
```

```python
# 셋의 교집합
set1 = {1,2,3}
set2 = {3,4,5}
intersection_set  =set1.intersection(set2)
intersection_set
```

```python
set1 &= set2
set1
```

```python
# 셋의 차집합
set1 = {1,2,3}
set2 = {3,4,5}
difference_set = set1.difference(set2)
difference_set
```

```python
set1 -= set2
set1
```
