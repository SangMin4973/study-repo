#### 딕셔너리
- 뮤터블 객체
- 키(key)와 값(value)쌍을 저장하는 해시 테이블

```python
my_dict = {}

my_dict["apple"] = 1
my_dict["banana"] = 2

my_dict
```

```python
# 딕셔너리 검색
key = "apple"
if key in my_dict:
    value = my_dict[key]
    print(f"{key}: {value}")
else:
    ("존재하지 않음")
```

```python
# 딕셔너리 수정
my_dict["banana"] = 4
my_dict
```

```python
# 딕셔너리 삭제
del my_dict["apple"]
my_dict
```
