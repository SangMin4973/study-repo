#### 앱실론
- 파이썬에서 부동 소수점을 표현할 때 이진법으로 표현하기 때문에 오차가 발생

```python
a = 0.1
b = 0.3

print(a * 3 == b)
print(f'3a = {3*a}, b = {b}')
```

```python
import sys
print(sys.float_info.epsilon)

a = 0.1 + 0.1 + 0.1
b = 0.3

print(a-b)

if abs(a-b) < sys.float_info.epsilon:
    print("a==b")
else:
    print("a!=b")
```
